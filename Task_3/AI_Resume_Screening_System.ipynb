```python
# Job Description
job_description = "Looking for Data Scientist with Python, SQL, Machine Learning, Pandas"
```

```python
# Resumes
resumes = {
    "Strong Candidate": "Python, SQL, Machine Learning, Pandas, 3 years experience",
    "Average Candidate": "Python, Excel, 1 year experience",
    "Weak Candidate": "Basic computer knowledge"
}
```

```python
# Step 1: Extract
def extract_chain(data):
    print("\n[Tracing] Step 1: Skill Extraction")
    return data["resume"].split(", ")
```

```python
# Step 2: Match
def match_chain(skills):
    print("[Tracing] Step 2: Matching Skills")
    required = ["Python", "SQL", "Machine Learning", "Pandas"]
    matched = [s for s in skills if s in required]
    missing = [s for s in required if s not in skills]
    return matched, missing
```

```python
# Step 3: Score
def score_chain(matched):
    print("[Tracing] Step 3: Scoring")
    return len(matched) * 25
```

```python
# Step 4: Explain
def explain_chain(score, matched, missing):
    print("[Tracing] Step 4: Explanation")
    
    if score >= 75:
        decision = "Strong Fit"
    elif score >= 50:
        decision = "Average Fit"
    else:
        decision = "Not Suitable"
    
    explanation = f"Matched: {matched}, Missing: {missing}"
    return decision, explanation
```

```python
# Main function
def run_pipeline(name, resume):
    print("\n==============================")
    print("Candidate:", name)
    
    skills = extract_chain({"resume": resume})
    print("Skills:", skills)
    
    matched, missing = match_chain(skills)
    print("Matched:", matched)
    print("Missing:", missing)
    
    score = score_chain(matched)
    print("Score:", score)
    
    decision, explanation = explain_chain(score, matched, missing)
    print("Decision:", decision)
    print("Explanation:", explanation)
```

```python
# Run all
for name, resume in resumes.items():
    run_pipeline(name, resume)
```
